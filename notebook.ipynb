```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def brownian_motion(ns):
    T = 1.0
    for n in ns:
        dt = T / n
        t = np.linspace(0, T, n)
        dW = np.random.normal(0, np.sqrt(dt), n)

        W = np.cumsum(dW)
        W = np.insert(W, 0, 0)[:-1]

        plt.plot(t, W)

    plt.show()
    
brownian_motion([1000])
```

```python
brownian_motion([100_000, 10_000, 1000, 100, 10])
```
