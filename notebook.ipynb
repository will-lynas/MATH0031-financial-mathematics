```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def brownian_motion(ns):
    T = 1.0
    for n in ns:
        dt = T / n
        t = np.linspace(0, T, n)
        dW = np.random.normal(0, np.sqrt(dt), n)

        W = np.cumsum(dW)
        W = np.insert(W, 0, 0)[:-1]

        plt.plot(t, W)

    plt.show()
    
brownian_motion([1000])
```

```python
brownian_motion([100_000, 10_000, 1000, 100, 10])
```

```python
def brownian_motion_2d(ns):
    T = 1.0
    for n in ns:
        dt = T / n
        dX = np.random.normal(0, np.sqrt(dt), n)
        dY = np.random.normal(0, np.sqrt(dt), n)

        X = np.cumsum(dX)
        Y = np.cumsum(dY)

        X = np.insert(X, 0, 0)[:-1]
        Y = np.insert(Y, 0, 0)[:-1]

        plt.plot(X, Y)
        plt.scatter(X[0], Y[0], color='red')

    plt.show()
    
brownian_motion_2d([1000])
```
