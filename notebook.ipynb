```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def brownian_motion(n):
    T = 1.0
    dt = T / n
    t = np.linspace(0, T, n)
    dW = np.random.normal(0, np.sqrt(dt), n)

    W = np.cumsum(dW)
    W = np.insert(W, 0, 0)[:-1]

    plt.plot(t, W)

    
brownian_motion(1000)
plt.show()
```

```python
brownian_motion(100_000)
brownian_motion(10_000)
brownian_motion(1000)
brownian_motion(100)
brownian_motion(10)
plt.show()
```

```python
def brownian_motion_2d(n, linewidth=1):
    T = 1.0
    dt = T / n
    dX = np.random.normal(0, np.sqrt(dt), n)
    dY = np.random.normal(0, np.sqrt(dt), n)

    X = np.cumsum(dX)
    Y = np.cumsum(dY)

    X = np.insert(X, 0, 0)[:-1]
    Y = np.insert(Y, 0, 0)[:-1]

    t = np.linspace(0, 1, len(X))
    colors = plt.cm.viridis(t)
    
    for i in range(len(X)-1):
        plt.plot(X[i:i+2], Y[i:i+2], color=colors[i], linewidth=linewidth)
    plt.scatter(X[0], Y[0], color='red')

    plt.show()
    
brownian_motion_2d(1000)
```

```python
brownian_motion_2d(10_000)
```

```python
brownian_motion_2d(100_000, linewidth=0.5)
```

```python
def gbm(mu=0.1, # Average return per unit time
        sigma=0.0, # Volatility (standard deviation per unit time)
        n=1000 # Number of steps
):
    S0 = 100
    T = 1.0
    dt = T / n
    t = np.linspace(0, 1.0, n)

    dW = np.random.normal(0, np.sqrt(dt), n)
    W = np.cumsum(dW)

    S = S0 * np.exp((mu - 0.5 * sigma**2) * t + sigma * W)

    plt.plot(t, S, label=f"={mu}, ={sigma}", linewidth=0.8)
```

```python
gbm()
gbm(sigma=0.01)
gbm(sigma=0.05)
gbm(sigma=0.1)
plt.legend()
plt.show()
```
